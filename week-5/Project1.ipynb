```python
import tkinter as tk
from tkinter import messagebox as msgbox
import random as rd

Employees = ['Mary Evans','Eyo Ishan','Durojaiye Dare','Adams Ali','Andy Ugwu','Stella Mark','Jane Akibo','Ago James','Michael Taiwo',
'David Jones','Nicole Akinde','Kosi Korso','Adele Martins','Emmanuel Ojo','Ajayi Fatima','Da-silva Justin', 'DT']
Tasks = ['Loading','Transporting','Reviewing Orders','Customer service','Delivering Items']
class Employee:
    def __init__(self,name):
        self.name = name
    def check_employee(self):
        if self.name in Employees:
            self.take_attendance()
            self.assign_task()
        else:
            self.refuse_access
    def take_attendance(self):
        print(f'Attendance recorded for:{self.name}')
    def assign_task(self):
        
        msgbox.showinfo('Access granted',f'Welcome {self.name}!!\n Your task for today is {rd.choice(Tasks)}')
    def refuse_access(self):
        msgbox.showerror('Access denied, User not found')

def on_submit():
    user_name = name.get().strip()
    if not user_name:
        msgbox.showwarning('Input error,\nPlease enter a name')
  
    session= Employee(user_name)
    session.check_employee()
    print ('Button clicked')
root = tk.Tk()
root.title("JANE & CO👜")
root.geometry("300x300")

tk.Label(root, text='JANE & CO👜', font=('monotype corsiva',25,'bold'),fg= 'Red').pack(pady= 10)
tk.Label(root, text= 'Enter your name',).pack(pady=15)
name = tk.Entry(root, width=30)
name.pack(pady=5)

submit = tk.Button(root, text= 'submit', command= on_submit, bg='Red',fg='Black',width=15 )
submit.pack(pady = 5)    

root.mainloop()
```
